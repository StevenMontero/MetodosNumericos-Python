```python
import numpy as np
def funx_s(x):
    return 7 + 14*(x**6)
    
def simpson(a,b,n):
    h = (b-a)/n
    XI_0 = funx_s(a) + funx_s(b)
    XI_1 = 0 # suma de f(x_2i-1)
    XI_2 = 0 # suma de f(x_2i)
    i = 1
    while i <= n-1:
        X = a + i*h
        if i%2 == 0:
            XI_2 = XI_2 + funx_s(X)
        else:
            XI_1 = XI_1 + funx_s(X)
        i = i+1
    XI = h*(XI_0+(2*XI_2)+(4*XI_1))/3
    return XI
```

```python
print(simpson(0,1,6))
```
