### 1. Utilice Python y Wolfram para

**a) Calcular** 
## $\sum_{k=0}^{950}\binom{1000}{k} \cdot a^k \cdot b^{1000-k}  $
donde $ a = 0,04 , b = 0.96$

**Recuerde que:**
## $ \binom{N}{K} = \frac{n!}{k!(n-k)!} $

### Python

````python
import numpy as np
from decimal import Decimal
sum_result = Decimal()
a = 0.04
b = 0.96
n = 1000
n_factorial = Decimal(np.math.factorial(n))
for k in range(950):
    k_factorial = Decimal(np.math.factorial(k))
    fun_result1 = n_factorial / (k_factorial * Decimal(np.math.factorial(n-k)))
    fun_result12 = Decimal(a**k) * Decimal(b**(1000-k))
    sum_result = sum_result + fun_result1 * fun_result12
print(sum_result) # resultado final de la sumatoria
print(sum_result.is_infinite()) # evalua si el resultado tiende a infinito
````

### Wolfram

**b) Use los mismos programas para calcular:**

## $\sum_{k=0}^{950} \left[ \prod_{j=1}^{k} \frac{1000 - k + j}{j} \cdot a\right] \cdot b^{1000-k}  $

### Python

````python
a = 0.04
b = 0.96
result_sum = 0.0
for k in range(950):
    result_prod = 1
    for j in range(1, k):
        result_prod *= ((1000-k+j)/j) * a
    result_sum += result_prod * b**(1000-k)
print(result_sum)
````

### Wolfram

```Wolfram

  prodResul = 1;
  For[j = 1, j <= k, j++,

   ];

  ];
```

### 2. Determine el error absoluto y relativo y el número de cifras significativas obtenidas para las siguientes aproximaciones:

````python
def error_calcultor(x,x_aprox):
    error_absoluto = np.absolute(x - x_aprox)
    error_relativo = error_absoluto/np.absolute(x)
    error_relativo_por = error_relativo*100
    cifras_signicativas = np.absolute(np.log(5) - np.log(error_relativo)) 
    return error_absoluto,error_relativo,error_relativo_por,cifras_signicativas
````

a) $\pi$ aproximado por $\frac{22}{7}$

````python
error_absoluto,error_relativo,error_por,cifras_signicativas = error_calcultor(np.pi , 22/7)
print('Error Absoluto:', error_absoluto)
print('Error Relativo:', error_relativo)
print('Error Porcentual:', error_por,'%')
print('Cifras signicativas:', int(cifras_signicativas))
````

````python
error_absoluto,error_relativo,error_por,cifras_signicativas = error_calcultor(0.3141592653589793 ,0.3142857142857143)
print('Error Absoluto:', error_absoluto)
print('Error Relativo:', error_relativo)
print('Error Porcentual:', error_por,'%')
print('Cifras signicativas:', int(cifras_signicativas))
````

b)$\frac{1}{3}$ aproximado por $0.333$ 

````python
error_absoluto,error_relativo,error_por,cifras_signicativas = error_calcultor(1/3,0.333)
print('Error Absoluto:', error_absoluto)
print('Error Relativo:', error_relativo)
print('Error Porcentual:', error_por,'%')
print('Cifras signicativas:', int(cifras_signicativas))
````

c) $\frac{1}{6}$ aproximado por $0.166$

````python
error_absoluto,error_relativo,error_por,cifras_signicativas = error_calcultor(1/6,0.166)
print('Error Absoluto:', error_absoluto)
print('Error Relativo:', error_relativo)
print('Error Porcentual:', error_por,'%')
print('Cifras signicativas:', int(cifras_signicativas))
````
